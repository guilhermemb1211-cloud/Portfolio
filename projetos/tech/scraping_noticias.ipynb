# **SCRAPING**
O scraping, ou web scraping, é uma técnica de extração de dados da web de forma automatizada. Basicamente, envolve o processo de acessar e coletar informações de sites da internet, geralmente por meio de um script ou programa, em vez de manualmente por um usuário. O scraping é usado para diversos fins, como coletar dados para análise, pesquisa, monitoramento de preços, comparação de produtos, entre outros. Porém, é importante observar as políticas de uso e os termos de serviço dos sites, já que nem todos permitem a extração automatizada de seus dados.

```python
#pip install google-play-scraper

import requests
from bs4 import BeautifulSoup

# URL do site de notícias
url = 'https://www.estadao.com.br/'

# Fazer uma requisição GET para obter o conteúdo da página
response = requests.get(url)
response.raise_for_status()  # Verifica se a requisição foi bem-sucedida
print(response)

# Parsear o conteúdo HTML com BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Encontrar elementos específicos, por exemplo, títulos de notícias
for headline in soup.find_all('h2', class_='headline'):
    print(headline.get_text().strip())
```
