**Léxicos**

Os léxicos são bases de conhecimento criadas manualmente ou de forma automatizada para atender um domínio específico (Linguística, Psicologia, Direito, etc.).
Vantagem: não há necessidade de rotular dados para treinamento.
Desvantagem:  léxicos são limitados a idioma e tamanho da base de conhecimento.

**SentilexPT**

É um léxico para o idioma Português possui 6531 adjetivos com informações de polaridade, alvo do sentimento e método de atribuição de polaridade.
tipo: PoS
Gênero: GN
alvo do sentimento: TG=HUM
polaridade: POL ( 1 ou -1 ou 0)
como a polaridade foi atribuida: ANOT

```python
#importar biblioteca numpy, pandas, io
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import io
```

```python
# Importar SentilexPT2
nomes=["palavra", "a", "polaridade", "b", "x","y"]
df = pd.read_csv("/content/drive/MyDrive/datasets/SentiLex-lem-PT02.csv", sep=';', names=nomes)
df.head(100)
```

```python
# apagar colunas desnecessárias deixar apenas coluna 0 (palavras) e coluna 2 (polaridade)
df = df.drop(['a','b','x','y'], axis=1)
df.head()
```

```python
#aplicar função split nas colunas Palavras e Polaridade
df['palavra']=df['palavra'].apply(lambda x: x.split('.'))
df['polaridade']=df['polaridade'].apply(lambda x: x.split('='))
df.head()
```

```python
# Limpar as listas dos campos palavra e polaridade deixando apenas o importante
df['palavra']=df['palavra'].apply(lambda x: x[0])
df['polaridade']=df['polaridade'].apply(lambda x: x[1])
df.head()
```

```python
# info()
df.info()
```

```python
#converter o campo polaridade para int
df['polaridade'] = df['polaridade'].astype(int)
```

```python
#criar um dicionário com as colunas palavra (key) polaridade (value) utilizndo a função zip.
dic = dict(zip(df['palavra'],df['polaridade']))
dic
```

```python
import string
pontuacao = string.punctuation
def Sentimento(frase,ic):
  frase.lower()
  frase = "".join([i for i in frase if i not in pontuacao])
  print(frase)
  frase = [i for i in frase.split() if i not in stopwords]
  print(frase)
  lista = [dic.get(i,0) for i in frase]
  print(lista)
  score = sum(lista)
  print(score)

frase = 'eu, Nilton, estou super feliz hoje, mas estou atrapalhado, entretanto, as coisas não estão boas, estão indo muito mal mas estou empolgado, com o trabalho dos meus Queridos alunos!!'
Sentimento(frase,dic)
```
